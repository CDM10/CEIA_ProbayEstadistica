```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def u2x(u):
    return np.ceil(u*6)
```

```python
u = np.random.rand(10000)
len(u)
```

```python
x = u2x(u)
x[:10]
```

```python
print([np.sum(x==k)/10000 for k in range(1,7)])
```

```python
def u2exp(u):
    return(-5*np.log(1-u))
```

```python
t = u2exp(u)
```

```python
import seaborn as sns
from scipy.stats import expon
sns.histplot(t, stat='density')
x= np.linspace(0,50,1000)
plt.plot(x, expon.pdf(x, scale=5), color='C1')
```
