```python
Ralpha=st.t.isf(0.05,49)
Ralpha
```

```python
Xbar=6.5
S   =1.5
n   =50
T   =(Xbar-6)/S*sqrt(n)
T
```

```python
print("rechazo H0" if T>Ralpha else "no rechazo H0")
```

# Ej 3

```python
# prior: elijo una prior conjugada de la binomial/bernoulli con media 0.7 y que sea concava
x = np.linspace(0,1,100)
mu =0.7
alphaplusbeta=10
alpha = mu*alphaplusbeta
beta=alphaplusbeta-alpha
plt.plot(x, st.beta.pdf(x,alpha,beta), color='C0')
print(f"alpha={alpha}, beta={beta}, mean={st.beta.mean(alpha,beta)}")
```

```python
# datos
y=5
n=7
```

```python
plt.plot(x, st.beta.pdf(x,alpha,beta), color='C0',label='a priori')
plt.plot(x, st.beta.pdf(x,12,5), color='C1',label='a posteriori')
plt.legend(loc='best')
plt.show()
```
