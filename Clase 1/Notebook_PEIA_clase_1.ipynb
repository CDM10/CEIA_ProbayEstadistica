# Ejercicio 2
$X\sim \mathcal{E}(1/5)$

$\mathbb{P}(X>2)$

```python
1-stats.expon.cdf(2, scale=5)
```

$\mathbb{P}(X>5|X>3)$

```python
(1-stats.expon.cdf(5, scale=5))/(1-stats.expon.cdf(3, scale=5))
```
