```python
y = np.linspace(-4,4,100)
```

```python
plt.plot(y,stats.norm.pdf(y), label='normal (0,1)')
plt.plot(y,stats.t.pdf(y,df=1), label='t df=1')
plt.plot(y,stats.t.pdf(y,df=5), label='t df=5')
plt.plot(y,stats.t.pdf(y,df=10), label='t df=10')
plt.plot(y,stats.t.pdf(y,df=25), label='t df=25')
plt.legend()
```

```python
n = 30 
```

```python
t_995=stats.t.ppf(.995,df=n-1)
t_995
```

```python
mu = 5
sigma = 2
```

```python
x = np.random.normal(size=n, loc=mu, scale=sigma)
```

```python
avg = x.mean()
sd = x.std()
```

```python
a = avg - t_995*sd/np.sqrt(n)
b = avg + t_995*sd/np.sqrt(n)
```

```python
print(a,b)
```

```python
print(a,b)
```

```python
chi_005=stats.chi2.ppf(0.005, df=n)
chi_995=stats.chi2.ppf(0.995, df=n)
```

```python
chi_005
```

```python
avg2 = (x**2).sum()
```

```python
a = (avg2-n*avg**2)/chi_995
b = (avg2-n*avg**2)/chi_005
```

```python
print(a,b)
```
