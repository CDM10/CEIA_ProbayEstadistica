```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
x = stats.expon.rvs(scale=2,size=10000)#lambda = 1/2
```

```python
plt.plot([x[:k].mean() for k in range(1,len(x))])
```

```python
u = np.random.uniform(size=(100,10000))
```

```python
n=5
un = u[:n].sum(axis=0)
plt.plot(np.linspace(-3.5,3.5,1000), stats.norm.pdf(x=np.linspace(-3.5,3.5,1000)),'r')
sns.histplot((un - un.mean())/un.std(), stat='density')
```

```python
b  = np.random.binomial(size=(1000,10000), n=5, p=0.1)
```

```python
n=1000
bn = b[:n].sum(axis=0)
plt.plot(np.linspace(-3.5,3.5,1000), stats.norm.pdf(x=np.linspace(-3.5,3.5,1000)),'r')
sns.histplot((bn - bn.mean())/bn.std(), stat='density')
```
